```python
import pandas as pd

# Load the Titanic dataset files
train_df = pd.read_csv('/content/train.csv')
test_df = pd.read_csv('/content/test.csv')
gender_submission_df = pd.read_csv('/content/gender_submission.csv')
```

```python
# Display the first few rows of the training data
train_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set plot styles
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# 1. Basic Information
basic_info = train_df.info()
```

```python
# 2. Descriptive Statistics
desc_stats = train_df.describe()
```

```python
# 3. Missing Values
missing_values = train_df.isnull().sum()
```

```python
# 4. Value Counts for Categorical Variables
sex_counts = train_df['Sex'].value_counts()
embarked_counts = train_df['Embarked'].value_counts()
pclass_counts = train_df['Pclass'].value_counts()
```

```python
# 5. Histograms
train_df['Age'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
train_df['Fare'].hist(bins=30, color='salmon', edgecolor='black')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
# 6. Boxplot for Age by Survival
sns.boxplot(x='Survived', y='Age', data=train_df, palette='Set2')
plt.title('Age Distribution by Survival')
plt.show()
```

```python
# 7. Countplot for Categorical Variables
sns.countplot(x='Sex', hue='Survived', data=train_df)
plt.title('Survival Count by Sex')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=train_df)
plt.title('Survival Count by Pclass')
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=train_df)
plt.title('Survival Count by Embarked Port')
plt.show()
```

```python
# 8. Heatmap for Correlation
numeric_df = train_df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# 9. Pairplot
sns.pairplot(numeric_df, hue='Survived', palette='husl')
plt.suptitle('Pairplot of Titanic Features', y=1.02)
plt.show()
```

```python
# Prepare summary
summary = {
    "Missing Values": missing_values,
    "Sex Counts": sex_counts,
    "Embarked Counts": embarked_counts,
    "Pclass Counts": pclass_counts,
    "Correlation Matrix": correlation_matrix
}
```

```python
# Fixing the pairplot by excluding 'Survived' from variables to be plotted directly
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
pairplot_df = train_df[features + ['Survived']].dropna()  # Drop rows with missing data for pairplot

# Convert 'Survived' to string for better hue representation
pairplot_df['Survived'] = pairplot_df['Survived'].astype(str)

# Now generate the corrected pairplot
sns.pairplot(pairplot_df, hue='Survived', palette='husl')
plt.suptitle('Corrected Pairplot of Titanic Features by Survival', y=1.02)
plt.show()
```

```python
# Ensure all data in the features are numeric and drop rows with any missing values
clean_df = train_df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']].dropna()

# Confirm all columns are numeric
clean_df = clean_df.astype({
    'Pclass': 'int',
    'Age': 'float',
    'SibSp': 'int',
    'Parch': 'int',
    'Fare': 'float',
    'Survived': 'int'
})

# Convert 'Survived' to string for hue
clean_df['Survived'] = clean_df['Survived'].astype(str)

# Generate the corrected pairplot
sns.pairplot(clean_df, hue='Survived', palette='husl')
plt.suptitle('Final Corrected Pairplot of Titanic Features by Survival', y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Clean the data
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']
clean_df = train_df[features].dropna()

# Convert to appropriate types
clean_df = clean_df.astype({
    'Pclass': 'int',
    'Age': 'float',
    'SibSp': 'int',
    'Parch': 'int',
    'Fare': 'float',
    'Survived': 'int'
})

# Set 'Survived' as string for hue
clean_df['Survived'] = clean_df['Survived'].astype(str)

# Plot
sns.pairplot(clean_df, hue='Survived', palette='husl')
plt.suptitle('Corrected Pairplot of Titanic Features by Survival', y=1.02)
plt.show()

```
